```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(0.5,1.6, 0.1)
```

```python
times = [28.008397, 
         21.751526, 
         18.539442, 
         14.389925,
         8.604313, 
         0.002406, 
         10.512507,
         19.771803,
         27.629700,
         35.464424, 
         41.666718]

xi = np.linspace(x[0], x[-1], 100)
ti = np.interp(xi,x,times,None)
```

```python
fig, ax = plt.subplots()
ax.plot(x, times, 'o', xi, ti, '.')
plt.xlabel('Scale Factor of Image')
plt.ylabel('Time in Seconds')
plt.title('Execution Time vs Scale Factor')
plt.show()
```

```python
fig.savefig("nornntimes.pdf", bbox_inches='tight')
```
